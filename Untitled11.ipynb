```python
# Anlise completa para 'ecommerce_estatistica.csv'
# Cole isto em um notebook Jupyter (ou Colab) e execute.
# Se o arquivo 'ecommerce_estatistica.csv' estiver no mesmo diretrio, ele ser usado.
# Caso contrrio, o script gera um dataset sinttico para demonstrao.

import os
import random
from datetime import datetime, timedelta
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# opcional: no Colab, descomente para mostrar arquivos salvos
# from google.colab import files

random_seed = 42
np.random.seed(random_seed)
random.seed(random_seed)

csv_path = "ecommerce_estatistica.csv"
use_df = None
notice = ""
if os.path.exists(csv_path):
    try:
        df = pd.read_csv(csv_path, parse_dates=True, infer_datetime_format=True)
        notice = f"Loaded '{csv_path}' from working directory."
        use_df = df.copy()
    except Exception as e:
        notice = f"Found '{csv_path}' but failed to read it: {e}. Generating sample dataset instead."
        use_df = None
else:
    notice = f"File '{csv_path}' not found in working directory. Generating a synthetic sample dataset to demonstrate the analysis."
    use_df = None

if use_df is None:
    # Build a synthetic ecommerce dataset
    n = 1000
    start_date = datetime(2024, 1, 1)
    order_dates = [start_date + timedelta(days=int(x)) for x in np.random.exponential(scale=180, size=n).cumsum() % 700]
    product_categories = ['Electronics', 'Clothing', 'Home', 'Books', 'Beauty', 'Toys']
    customers = [f"user{np.random.randint(1,301)}@example.com" for _ in range(n)]
    prices = np.round(np.abs(np.random.normal(loc=50, scale=40, size=n)) + 1, 2)
    quantities = np.random.poisson(lam=1.6, size=n) + 1
    revenue = np.round(prices * quantities, 2)
    ratings = np.clip(np.round(np.random.normal(4.0, 0.8, size=n), 1), 1.0, 5.0)
    categories = [random.choice(product_categories) for _ in range(n)]
    order_ids = [f"ORD{100000+i}" for i in range(n)]
    df = pd.DataFrame({
        'order_id': order_ids,
        'customer_email': customers,
        'order_date': order_dates,
        'product_category': categories,
        'price': prices,
        'quantity': quantities,
        'revenue': revenue,
        'rating': ratings
    })
    use_df = df.copy()

# Try coerce order_date to datetime
if 'order_date' in use_df.columns:
    use_df['order_date'] = pd.to_datetime(use_df['order_date'], errors='coerce')

# Overview
rows, cols = use_df.shape
print(notice)
print(f"Dataset contains {rows} rows and {cols} columns.")
print("Columns:", ", ".join(use_df.columns))
display(use_df.head(8))

# Create output directory (onde os grficos sero salvos)
out_dir = "ecommerce_analysis_outputs"
os.makedirs(out_dir, exist_ok=True)

# 1) Histograma (preo)
plt.figure(figsize=(8,5))
use_df['price'].hist(bins=30)
plt.title("Histograma de Preos - Distribuio dos preos dos itens")
plt.xlabel("Preo (unidade)")
plt.ylabel("Contagem")
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "histograma_preco.png"))
plt.show()
plt.close()

# 2) Disperso (price vs quantity)
plt.figure(figsize=(8,5))
plt.scatter(use_df['price'], use_df['quantity'])
plt.title("Disperso: Preo vs Quantidade vendida")
plt.xlabel("Preo (unidade)")
plt.ylabel("Quantidade")
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "dispersao_preco_quantidade.png"))
plt.show()
plt.close()

# 3) Mapa de calor (correlao numrica) - sem seaborn (usando imshow)
numeric_cols = use_df.select_dtypes(include=[np.number]).columns.tolist()
corr = use_df[numeric_cols].corr()
plt.figure(figsize=(7,6))
im = plt.imshow(corr.values, aspect='auto', interpolation='nearest')
plt.colorbar(im)
plt.xticks(range(len(numeric_cols)), numeric_cols, rotation=45, ha='right')
plt.yticks(range(len(numeric_cols)), numeric_cols)
plt.title("Mapa de Calor: Matriz de correlao (numrica)")
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "mapa_calor_correlacao.png"))
plt.show()
plt.close()

# 4) Grfico de barra (receita por categoria)
rev_by_cat = use_df.groupby('product_category')['revenue'].sum().sort_values(ascending=False)
plt.figure(figsize=(8,5))
rev_by_cat.plot(kind='bar')
plt.title("Receita por Categoria de Produto")
plt.xlabel("Categoria de Produto")
plt.ylabel("Receita Total")
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "barra_receita_categoria.png"))
plt.show()
plt.close()

# 5) Grfico de pizza (participao por categoria - receita%)
plt.figure(figsize=(6,6))
shares = rev_by_cat / rev_by_cat.sum()
plt.pie(shares, labels=shares.index, autopct='%1.1f%%')
plt.title("Participao (%) da Receita por Categoria")
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "pizza_participacao_categoria.png"))
plt.show()
plt.close()

# 6) Densidade (price)
plt.figure(figsize=(8,5))
use_df['price'].plot(kind='density')
plt.title("Densidade de Preos")
plt.xlabel("Preo (unidade)")
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "densidade_preco.png"))
plt.show()
plt.close()

# 7) Regresso (linha de tendncia preo -> receita)
x = use_df['price'].values
y = use_df['revenue'].values
mask = np.isfinite(x) & np.isfinite(y)
coeffs = np.polyfit(x[mask], y[mask], deg=1)
slope, intercept = coeffs
reg_y = slope * x + intercept
plt.figure(figsize=(8,5))
plt.scatter(x, y, alpha=0.6)
plt.plot(x, reg_y, linewidth=2)
plt.title("Regresso Linear: Preo -> Receita (linha de tendncia)")
plt.xlabel("Preo (unidade)")
plt.ylabel("Receita")
plt.annotate(f"y = {slope:.3f}x + {intercept:.2f}", xy=(0.05, 0.95), xycoords='axes fraction', ha='left', va='top')
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "regressao_preco_receita.png"))
plt.show()
plt.close()

# Estatsticas resumidas
display(use_df.describe(include='all').T)

print(f"Grficos salvos na pasta: {out_dir}")
# Anlise completa para 'ecommerce_estatistica.csv'
# Cole isto em um notebook Jupyter (ou Colab) e execute.
# Se o arquivo 'ecommerce_estatistica.csv' estiver no mesmo diretrio, ele ser usado.
# Caso contrrio, o script gera um dataset sinttico para demonstrao.

import os
import random
from datetime import datetime, timedelta
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# opcional: no Colab, descomente para mostrar arquivos salvos
# from google.colab import files

random_seed = 42
np.random.seed(random_seed)
random.seed(random_seed)

csv_path = "ecommerce_estatistica.csv"
use_df = None
notice = ""
if os.path.exists(csv_path):
    try:
        df = pd.read_csv(csv_path, parse_dates=True, infer_datetime_format=True)
        notice = f"Loaded '{csv_path}' from working directory."
        use_df = df.copy()
    except Exception as e:
        notice = f"Found '{csv_path}' but failed to read it: {e}. Generating sample dataset instead."
        use_df = None
else:
    notice = f"File '{csv_path}' not found in working directory. Generating a synthetic sample dataset to demonstrate the analysis."
    use_df = None

if use_df is None:
    # Build a synthetic ecommerce dataset
    n = 1000
    start_date = datetime(2024, 1, 1)
    order_dates = [start_date + timedelta(days=int(x)) for x in np.random.exponential(scale=180, size=n).cumsum() % 700]
    product_categories = ['Electronics', 'Clothing', 'Home', 'Books', 'Beauty', 'Toys']
    customers = [f"user{np.random.randint(1,301)}@example.com" for _ in range(n)]
    prices = np.round(np.abs(np.random.normal(loc=50, scale=40, size=n)) + 1, 2)
    quantities = np.random.poisson(lam=1.6, size=n) + 1
    revenue = np.round(prices * quantities, 2)
    ratings = np.clip(np.round(np.random.normal(4.0, 0.8, size=n), 1), 1.0, 5.0)
    categories = [random.choice(product_categories) for _ in range(n)]
    order_ids = [f"ORD{100000+i}" for i in range(n)]
    df = pd.DataFrame({
        'order_id': order_ids,
        'customer_email': customers,
        'order_date': order_dates,
        'product_category': categories,
        'price': prices,
        'quantity': quantities,
        'revenue': revenue,
        'rating': ratings
    })
    use_df = df.copy()

# Try coerce order_date to datetime
if 'order_date' in use_df.columns:
    use_df['order_date'] = pd.to_datetime(use_df['order_date'], errors='coerce')

# Overview
rows, cols = use_df.shape
print(notice)
print(f"Dataset contains {rows} rows and {cols} columns.")
print("Columns:", ", ".join(use_df.columns))
display(use_df.head(8))

# Create output directory (onde os grficos sero salvos)
out_dir = "ecommerce_analysis_outputs"
os.makedirs(out_dir, exist_ok=True)

# 1) Histograma (preo)
plt.figure(figsize=(8,5))
use_df['price'].hist(bins=30)
plt.title("Histograma de Preos - Distribuio dos preos dos itens")
plt.xlabel("Preo (unidade)")
plt.ylabel("Contagem")
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "histograma_preco.png"))
plt.show()
plt.close()

# 2) Disperso (price vs quantity)
plt.figure(figsize=(8,5))
plt.scatter(use_df['price'], use_df['quantity'])
plt.title("Disperso: Preo vs Quantidade vendida")
plt.xlabel("Preo (unidade)")
plt.ylabel("Quantidade")
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "dispersao_preco_quantidade.png"))
plt.show()
plt.close()

# 3) Mapa de calor (correlao numrica) - sem seaborn (usando imshow)
numeric_cols = use_df.select_dtypes(include=[np.number]).columns.tolist()
corr = use_df[numeric_cols].corr()
plt.figure(figsize=(7,6))
im = plt.imshow(corr.values, aspect='auto', interpolation='nearest')
plt.colorbar(im)
plt.xticks(range(len(numeric_cols)), numeric_cols, rotation=45, ha='right')
plt.yticks(range(len(numeric_cols)), numeric_cols)
plt.title("Mapa de Calor: Matriz de correlao (numrica)")
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "mapa_calor_correlacao.png"))
plt.show()
plt.close()

# 4) Grfico de barra (receita por categoria)
rev_by_cat = use_df.groupby('product_category')['revenue'].sum().sort_values(ascending=False)
plt.figure(figsize=(8,5))
rev_by_cat.plot(kind='bar')
plt.title("Receita por Categoria de Produto")
plt.xlabel("Categoria de Produto")
plt.ylabel("Receita Total")
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "barra_receita_categoria.png"))
plt.show()
plt.close()

# 5) Grfico de pizza (participao por categoria - receita%)
plt.figure(figsize=(6,6))
shares = rev_by_cat / rev_by_cat.sum()
plt.pie(shares, labels=shares.index, autopct='%1.1f%%')
plt.title("Participao (%) da Receita por Categoria")
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "pizza_participacao_categoria.png"))
plt.show()
plt.close()

# 6) Densidade (price)
plt.figure(figsize=(8,5))
use_df['price'].plot(kind='density')
plt.title("Densidade de Preos")
plt.xlabel("Preo (unidade)")
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "densidade_preco.png"))
plt.show()
plt.close()

# 7) Regresso (linha de tendncia preo -> receita)
x = use_df['price'].values
y = use_df['revenue'].values
mask = np.isfinite(x) & np.isfinite(y)
coeffs = np.polyfit(x[mask], y[mask], deg=1)
slope, intercept = coeffs
reg_y = slope * x + intercept
plt.figure(figsize=(8,5))
plt.scatter(x, y, alpha=0.6)
plt.plot(x, reg_y, linewidth=2)
plt.title("Regresso Linear: Preo -> Receita (linha de tendncia)")
plt.xlabel("Preo (unidade)")
plt.ylabel("Receita")
plt.annotate(f"y = {slope:.3f}x + {intercept:.2f}", xy=(0.05, 0.95), xycoords='axes fraction', ha='left', va='top')
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "regressao_preco_receita.png"))
plt.show()
plt.close()

# Estatsticas resumidas
display(use_df.describe(include='all').T)

print(f"Grficos salvos na pasta: {out_dir}")
# Anlise completa para 'ecommerce_estatistica.csv'
# Cole isto em um notebook Jupyter (ou Colab) e execute.
# Se o arquivo 'ecommerce_estatistica.csv' estiver no mesmo diretrio, ele ser usado.
# Caso contrrio, o script gera um dataset sinttico para demonstrao.

import os
import random
from datetime import datetime, timedelta
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# opcional: no Colab, descomente para mostrar arquivos salvos
# from google.colab import files

random_seed = 42
np.random.seed(random_seed)
random.seed(random_seed)

csv_path = "ecommerce_estatistica.csv"
use_df = None
notice = ""
if os.path.exists(csv_path):
    try:
        df = pd.read_csv(csv_path, parse_dates=True, infer_datetime_format=True)
        notice = f"Loaded '{csv_path}' from working directory."
        use_df = df.copy()
    except Exception as e:
        notice = f"Found '{csv_path}' but failed to read it: {e}. Generating sample dataset instead."
        use_df = None
else:
    notice = f"File '{csv_path}' not found in working directory. Generating a synthetic sample dataset to demonstrate the analysis."
    use_df = None

if use_df is None:
    # Build a synthetic ecommerce dataset
    n = 1000
    start_date = datetime(2024, 1, 1)
    order_dates = [start_date + timedelta(days=int(x)) for x in np.random.exponential(scale=180, size=n).cumsum() % 700]
    product_categories = ['Electronics', 'Clothing', 'Home', 'Books', 'Beauty', 'Toys']
    customers = [f"user{np.random.randint(1,301)}@example.com" for _ in range(n)]
    prices = np.round(np.abs(np.random.normal(loc=50, scale=40, size=n)) + 1, 2)
    quantities = np.random.poisson(lam=1.6, size=n) + 1
    revenue = np.round(prices * quantities, 2)
    ratings = np.clip(np.round(np.random.normal(4.0, 0.8, size=n), 1), 1.0, 5.0)
    categories = [random.choice(product_categories) for _ in range(n)]
    order_ids = [f"ORD{100000+i}" for i in range(n)]
    df = pd.DataFrame({
        'order_id': order_ids,
        'customer_email': customers,
        'order_date': order_dates,
        'product_category': categories,
        'price': prices,
        'quantity': quantities,
        'revenue': revenue,
        'rating': ratings
    })
    use_df = df.copy()

# Try coerce order_date to datetime
if 'order_date' in use_df.columns:
    use_df['order_date'] = pd.to_datetime(use_df['order_date'], errors='coerce')

# Overview
rows, cols = use_df.shape
print(notice)
print(f"Dataset contains {rows} rows and {cols} columns.")
print("Columns:", ", ".join(use_df.columns))
display(use_df.head(8))

# Create output directory (onde os grficos sero salvos)
out_dir = "ecommerce_analysis_outputs"
os.makedirs(out_dir, exist_ok=True)

# 1) Histograma (preo)
plt.figure(figsize=(8,5))
use_df['price'].hist(bins=30)
plt.title("Histograma de Preos - Distribuio dos preos dos itens")
plt.xlabel("Preo (unidade)")
plt.ylabel("Contagem")
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "histograma_preco.png"))
plt.show()
plt.close()

# 2) Disperso (price vs quantity)
plt.figure(figsize=(8,5))
plt.scatter(use_df['price'], use_df['quantity'])
plt.title("Disperso: Preo vs Quantidade vendida")
plt.xlabel("Preo (unidade)")
plt.ylabel("Quantidade")
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "dispersao_preco_quantidade.png"))
plt.show()
plt.close()

# 3) Mapa de calor (correlao numrica) - sem seaborn (usando imshow)
numeric_cols = use_df.select_dtypes(include=[np.number]).columns.tolist()
corr = use_df[numeric_cols].corr()
plt.figure(figsize=(7,6))
im = plt.imshow(corr.values, aspect='auto', interpolation='nearest')
plt.colorbar(im)
plt.xticks(range(len(numeric_cols)), numeric_cols, rotation=45, ha='right')
plt.yticks(range(len(numeric_cols)), numeric_cols)
plt.title("Mapa de Calor: Matriz de correlao (numrica)")
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "mapa_calor_correlacao.png"))
plt.show()
plt.close()

# 4) Grfico de barra (receita por categoria)
rev_by_cat = use_df.groupby('product_category')['revenue'].sum().sort_values(ascending=False)
plt.figure(figsize=(8,5))
rev_by_cat.plot(kind='bar')
plt.title("Receita por Categoria de Produto")
plt.xlabel("Categoria de Produto")
plt.ylabel("Receita Total")
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "barra_receita_categoria.png"))
plt.show()
plt.close()

# 5) Grfico de pizza (participao por categoria - receita%)
plt.figure(figsize=(6,6))
shares = rev_by_cat / rev_by_cat.sum()
plt.pie(shares, labels=shares.index, autopct='%1.1f%%')
plt.title("Participao (%) da Receita por Categoria")
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "pizza_participacao_categoria.png"))
plt.show()
plt.close()

# 6) Densidade (price)
plt.figure(figsize=(8,5))
use_df['price'].plot(kind='density')
plt.title("Densidade de Preos")
plt.xlabel("Preo (unidade)")
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "densidade_preco.png"))
plt.show()
plt.close()

# 7) Regresso (linha de tendncia preo -> receita)
x = use_df['price'].values
y = use_df['revenue'].values
mask = np.isfinite(x) & np.isfinite(y)
coeffs = np.polyfit(x[mask], y[mask], deg=1)
slope, intercept = coeffs
reg_y = slope * x + intercept
plt.figure(figsize=(8,5))
plt.scatter(x, y, alpha=0.6)
plt.plot(x, reg_y, linewidth=2)
plt.title("Regresso Linear: Preo -> Receita (linha de tendncia)")
plt.xlabel("Preo (unidade)")
plt.ylabel("Receita")
plt.annotate(f"y = {slope:.3f}x + {intercept:.2f}", xy=(0.05, 0.95), xycoords='axes fraction', ha='left', va='top')
plt.tight_layout()
plt.savefig(os.path.join(out_dir, "regressao_preco_receita.png"))
plt.show()
plt.close()

# Estatsticas resumidas
display(use_df.describe(include='all').T)

print(f"Grficos salvos na pasta: {out_dir}")
```
